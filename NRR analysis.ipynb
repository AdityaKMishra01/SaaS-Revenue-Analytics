# Net Revenue Retention (NRR) Analysis

## Objective
Measure revenue retained from existing customers after accounting for expansion, contraction, and churn.

## Input Files
- monthly_mrr.csv
- monthly_expansion.csv
- monthly_contraction.csv
- monthly_churn.csv

## Output
- nrr_monthly.csv

```python
import pandas as pd
```

```python
monthly_mrr = pd.read_csv('monthly_mrr.csv')
monthly_expansion = pd.read_csv('monthly_expansion.csv')
monthly_contraction = pd.read_csv('monthly_contraction.csv')
monthly_churn = pd.read_csv('monthly_churn.csv')
```

```python
monthly_mrr['starting_mrr'] = monthly_mrr['mrr_amount'].shift(1) 
```

```python
monthly_mrr
```

```python
nrr_df = monthly_mrr.copy()
```

```python
nrr_df=pd.merge(nrr_df,monthly_expansion,on='year_and_month',how='left')
```

```python
nrr_df=pd.merge(nrr_df,monthly_contraction,on='year_and_month',how='left')
```

```python
nrr_df=pd.merge(nrr_df,monthly_churn,on='year_and_month',how='left')
```

```python
nrr_df = nrr_df[nrr_df['starting_mrr'] > 0]
```

```python
nrr_df = nrr_df.rename(columns = {'contraction_amount_x':'contraction_amount'})
nrr_df
```

```python
nrr_df = nrr_df.fillna(0)
```

```python
nrr_df.isnull().sum()
```

```python
nrr_df.columns
```

```python
nrr_df['ending_mrr'] = (
    nrr_df['starting_mrr']
    + nrr_df['expansion_amount']
    - nrr_df['contraction_amount']
    - nrr_df['churn_amount']
)
```

```python
nrr_df['nrr_percent'] = (
    nrr_df['ending_mrr'] / nrr_df['starting_mrr']
) * 100
```

```python
nrr_df
```

```python
nrr_df.to_csv('nrr_df.csv',index=False)
```

# Net Revenue Retention (NRR) Insights

## Business Interpretation

Net Revenue Retention (NRR) measures how much recurring revenue is retained from existing customers after accounting for expansion, contraction, and churn.

An NRR value above 100% indicates that revenue growth from existing customers exceeds revenue lost through downgrades and churn. An NRR value below 100% indicates net revenue loss from the existing customer base.

## Key Findings

- Expansion MRR contributed positively to revenue growth by increasing revenue from existing customers.
- Contraction MRR reduced revenue when customers downgraded plans or reduced usage.
- Churn MRR represented the largest source of revenue leakage because customers completely stopped generating revenue.
- NRR provides a consolidated view of customer retention and revenue growth performance.
- Sustaining NRR above 100% is a strong indicator of a healthy SaaS business model.

## Conclusion

The NRR analysis demonstrates how revenue changes over time due to customer behavior. By monitoring expansion, contraction, and churn, businesses can identify growth opportunities and reduce revenue leakage, ultimately improving long-term recurring revenue performance.
